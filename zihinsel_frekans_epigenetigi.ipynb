```python
import numpy as np
import matplotlib.pyplot as plt

def zihinsel_frekans_epigenetigi(x, r, e, t):
    """
    Zihinsel Frekans Epigenetii fonksiyonu
    x: zihinsel frekans girii (rnein duygu-dnce dzeyi)
    r: rezonans katsays (iletiim gc)
    e: epigenetik duyarllk (genetik bilgiye etki dzeyi)
    t: evresel/topografik etkiler (haritac balam)
    """
    y = (x * r) + (e * (x ** 2)) - (0.1 * t)
    return y

def generate_signal_data(length=100, noise_level=0.05):
    """
    Zihinsel frekanslar temsil eden yapay bir sinyal oluturur
    """
    time = np.linspace(0, 10, length)
    frequency = 0.5 * np.sin(time) + 0.5  # Duygusal frekanslar [0, 1] aras
    noise = np.random.normal(0, noise_level, size=length)
    signal = frequency + noise
    return time, signal

def main():
    # Sabit parametreler
    rezonans = 1.2
    epigenetik = 0.6
    topografya = 0.4

    # Zaman serisi verisi ret
    time, signal = generate_signal_data()

    # Model ktsn hesapla
    etkiler = [zihinsel_frekans_epigenetigi(x, rezonans, epigenetik, topografya) for x in signal]

    # Grselletir
    plt.figure(figsize=(10, 5))
    plt.plot(time, signal, label='Zihinsel Frekans Sinyali', linestyle='--')
    plt.plot(time, etkiler, label='Epigenetik Etki kts', linewidth=2)
    plt.xlabel('Zaman')
    plt.ylabel('Frekans / Etki')
    plt.title('Zihinsel Frekans Epigenetii - Zaman Serisi Simlasyonu')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
