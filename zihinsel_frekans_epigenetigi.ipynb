```python
import numpy as np
import matplotlib.pyplot as plt

def zihinsel_frekans_epigenetigi(x, r, e, t):
    """
    Zihinsel Frekans Epigenetiği fonksiyonu
    x: zihinsel frekans girişi (örneğin duygu-düşünce düzeyi)
    r: rezonans katsayısı (iletişim gücü)
    e: epigenetik duyarlılık (genetik bilgiye etki düzeyi)
    t: çevresel/topografik etkiler (haritacı bağlamı)
    """
    y = (x * r) + (e * (x ** 2)) - (0.1 * t)
    return y

def generate_signal_data(length=100, noise_level=0.05):
    """
    Zihinsel frekansları temsil eden yapay bir sinyal oluşturur
    """
    time = np.linspace(0, 10, length)
    frequency = 0.5 * np.sin(time) + 0.5  # Duygusal frekanslar [0, 1] arası
    noise = np.random.normal(0, noise_level, size=length)
    signal = frequency + noise
    return time, signal

def main():
    # Sabit parametreler
    rezonans = 1.2
    epigenetik = 0.6
    topografya = 0.4

    # Zaman serisi verisi üret
    time, signal = generate_signal_data()

    # Model çıktısını hesapla
    etkiler = [zihinsel_frekans_epigenetigi(x, rezonans, epigenetik, topografya) for x in signal]

    # Görselleştir
    plt.figure(figsize=(10, 5))
    plt.plot(time, signal, label='Zihinsel Frekans Sinyali', linestyle='--')
    plt.plot(time, etkiler, label='Epigenetik Etki Çıktısı', linewidth=2)
    plt.xlabel('Zaman')
    plt.ylabel('Frekans / Etki')
    plt.title('Zihinsel Frekans Epigenetiği - Zaman Serisi Simülasyonu')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
