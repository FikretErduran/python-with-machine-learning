```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# LinarRegression formül=y= mx+b

dosya= pd.read_csv('Student.csv')

dosya.head(10)

y=dosya[['Marks']]
x=dosya[['number_courses','time_study']]


#sklearn içerisindeli linear_model.LinearRegression()
linear_model=LinearRegression() # linear_model adındaki değişkene-->LinearRegression modelini atadık

model=linear_model.fit(x,y) # x (number_courses)  ve y (time_study) lis değişkenleri
#makine öğrenimi için eklemek

model.predict([[2,10]]) # 4. sınıfta 18 saat çalışılınca alınacak puan

dosya['Marks'].max() #maksimum değer

model.score(x,y) # skor tahmini
```
