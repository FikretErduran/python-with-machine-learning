```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# üî∏ √ñrnek veriler: uzaklƒ±k (milyon km) ve g√∂zlenen ƒ±≈üƒ±k parlaklƒ±ƒüƒ± (g√∂reli birim)
# Ger√ßek astronomik veriler deƒüildir, sadece √∂rnek sim√ºlasyon i√ßin
distance = np.array([350, 360, 370, 380, 390, 400]).reshape(-1, 1)  # Ay uzaklƒ±ƒüƒ± (bin km)
brightness = np.array([0.95, 0.89, 0.85, 0.80, 0.76, 0.72])        # I≈üƒ±k parlaklƒ±ƒüƒ±

# üîπ Model olu≈ütur
model = LinearRegression()
model.fit(distance, brightness)

# üî∏ Tahmin yap
future_distance = np.array([410, 420, 430]).reshape(-1, 1)
predicted_brightness = model.predict(future_distance)

# üîπ Grafikle g√∂ster
plt.scatter(distance, brightness, color='yellow', label='Ger√ßek veriler')
plt.plot(distance, model.predict(distance), color='blue', label='Regresyon √ßizgisi')
plt.scatter(future_distance, predicted_brightness, color='red', label='Tahmin')
plt.xlabel('Uzaklƒ±k (bin km)')
plt.ylabel('Parlaklƒ±k (g√∂reli birim)')
plt.title('Ay Uzaklƒ±ƒüƒ± ile I≈üƒ±k Parlaklƒ±ƒüƒ± Arasƒ±ndaki ƒ∞li≈üki')
plt.legend()
plt.grid(True)
plt.show()

# üîç Model detaylarƒ±
print("Regresyon katsayƒ±sƒ± (eƒüim):", model.coef_[0])
print("Sabit terim (bias):", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# üî∏ Zaman verisi (dakika cinsinden)
minutes = np.arange(1, 16).reshape(-1, 1)  # 1'den 15'e kadar

# üîπ Her dakikada 0.15 grad kayma olmu≈ü
angle_shift = np.array([0.15 * i for i in range(1, 16)])  # toplam a√ßƒ± kaymasƒ± (grad)

# üî∏ Regresyon modeli
model = LinearRegression()
model.fit(minutes, angle_shift)

# üîπ Tahmin (√∂rneƒüin 20. dakikadaki a√ßƒ± kaymasƒ±)
future_minutes = np.array([16, 17, 18, 19, 20]).reshape(-1, 1)
predicted_angles = model.predict(future_minutes)

# üî∏ Grafik
plt.plot(minutes, angle_shift, 'o', label='G√∂zlem Verileri', color='orange')
plt.plot(minutes, model.predict(minutes), label='Regresyon √áizgisi', color='blue')
plt.plot(future_minutes, predicted_angles, 'x', label='Tahmin', color='red')
plt.xlabel('Zaman (dakika)')
plt.ylabel('A√ßƒ± Kaymasƒ± (grad)')
plt.title('Zamana Baƒülƒ± Ay‚Äôƒ±n G√∂r√ºn√ºr A√ßƒ± Kaymasƒ±')
plt.grid(True)
plt.legend()
plt.show()

# üîç Model bilgisi
print("Regresyon eƒüimi (grad/dk):", model.coef_[0])
print("Sabit terim (ba≈ülangƒ±√ß a√ßƒ±sƒ±):", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sabitler
earth_radius_km = 6371         # D√ºnya yarƒ±√ßapƒ±
true_angular_velocity_deg = 0.25  # D√ºnya'nƒ±n ger√ßek d√∂nme hƒ±zƒ± (derece/dakika)
measured_velocity_deg = 0.135     # √ñl√ß√ºlen a√ßƒ± hƒ±zƒ± (senin verin)

# 15 dakikalƒ±k g√∂zlem
minutes = np.arange(1, 16)
true_angles = true_angular_velocity_deg * minutes
measured_angles = measured_velocity_deg * minutes

# Uzaklƒ±k tahmini (basit orantƒ±: tan(a) ~ a k√º√ß√ºk a√ßƒ±larda)
# Sabit bir z tanƒ±mlayalƒ±m ve g√∂reli uzaklƒ±k farkƒ±nƒ± g√∂relim
# tan(theta) = kar≈üƒ± / kom≈üu -> kom≈üu = kar≈üƒ± / tan(theta)
# Burada kar≈üƒ± uzunluk sabit, √∂rneƒüin totalstation‚Äôƒ±n g√∂r√º≈ü √ßizgisi
Z_fixed = 1  # sabit y√ºkseklik (g√∂zlem d√ºzlemi)
true_distances = Z_fixed / np.tan(np.radians(true_angles))
measured_distances = Z_fixed / np.tan(np.radians(measured_angles))

# Grafik
plt.plot(minutes, true_distances, label='Ger√ßek Uzaklƒ±k Eƒürisi', color='green')
plt.plot(minutes, measured_distances, label='√ñl√ß√ºlen Eƒüri (senin veri)', color='orange', linestyle='--')
plt.xlabel("Zaman (dakika)")
plt.ylabel("G√∂reli Uzaklƒ±k (birim)")
plt.title("D√ºnya-Ay Uzaklƒ±ƒüƒ±na Kar≈üƒ± A√ßƒ±sal Kayma")
plt.legend()
plt.grid(True)
plt.show()
```
