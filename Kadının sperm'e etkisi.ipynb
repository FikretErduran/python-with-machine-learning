```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parametreler
kadinin_enerjisi = np.linspace(0, 1, 50)  # E: Kadn enerjisi
erkegin_biyokimya = np.linspace(0, 1, 50)  # B: Erkek biyokimyas

E, B = np.meshgrid(kadinin_enerjisi, erkegin_biyokimya)

# DNA Btnl (D) basit model: rezonans etkisi arpanlarla
# D = E * B + 0.2*E*B*(1 - E*B)  -> hem dorusal hem rezonans etkisi
D = E * B + 0.3 * np.sin(np.pi * E * B)

# Grafik
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(E, B, D, cmap='viridis', edgecolor='k', alpha=0.8)

ax.set_xlabel("Kadnn Enerjisi (E)")
ax.set_ylabel("Erkein Biyokimyas (B)")
ax.set_zlabel("Sperm DNA Epigenetik Btnl (D)")
ax.set_title("Kadn Enerjisi ve Erkek Biyokimyas Etkisi ile DNA Btnl Rezonans")

plt.show()
```
