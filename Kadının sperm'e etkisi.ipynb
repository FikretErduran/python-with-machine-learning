```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parametreler
kadinin_enerjisi = np.linspace(0, 1, 50)  # E: Kadın enerjisi
erkegin_biyokimya = np.linspace(0, 1, 50)  # B: Erkek biyokimyası

E, B = np.meshgrid(kadinin_enerjisi, erkegin_biyokimya)

# DNA Bütünlüğü (D) basit model: rezonans etkisi çarpanlarla
# D = E * B + 0.2*E*B*(1 - E*B)  -> hem doğrusal hem rezonans etkisi
D = E * B + 0.3 * np.sin(np.pi * E * B)

# Grafik
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(E, B, D, cmap='viridis', edgecolor='k', alpha=0.8)

ax.set_xlabel("Kadının Enerjisi (E)")
ax.set_ylabel("Erkeğin Biyokimyası (B)")
ax.set_zlabel("Sperm DNA Epigenetik Bütünlüğü (D)")
ax.set_title("Kadın Enerjisi ve Erkek Biyokimyası Etkisi ile DNA Bütünlüğü Rezonansı")

plt.show()
```
