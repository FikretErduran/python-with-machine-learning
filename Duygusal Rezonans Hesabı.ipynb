```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Duygusal yoğunluk fonksiyonu (kullanıcı burayı özelleştirebilir)
def f(t, a=10, b=0.5, c=5):
    return a * np.sin(b * t) + c

# İntegral hesabı
def hesapla_duygusal_enerji(t0, t1, a=10, b=0.5, c=5):
    integral, _ = quad(f, t0, t1, args=(a, b, c))
    return integral

# Grafik çizimi
def grafik_ciz(t0, t1, a=10, b=0.5, c=5):
    t = np.linspace(t0, t1, 500)
    y = f(t, a, b, c)

    plt.figure(figsize=(10, 5))
    plt.plot(t, y, label="Duygusal Yoğunluk f(t)", color="crimson")
    plt.fill_between(t, y, alpha=0.3, color="salmon")
    plt.title("Zamana Bağlı Duygusal Rezonans")
    plt.xlabel("Zaman (t)")
    plt.ylabel("Yoğunluk")
    plt.grid(True)
    plt.legend()
    plt.show()

# Kullanıcı girişleri
t0 = 2021       # başlangıç zamanı
t1 = 2025      # bitiş zamanı
a, b, c = 10, 0.5, 5  # duygu eğrisi parametreleri

# Hesaplama ve çıktı
enerji = hesapla_duygusal_enerji(t0, t1, a, b, c)
print(f"{t0} ile {t1} zamanı arasındaki toplam duygusal enerji: {enerji:.2f} birim")
grafik_ciz(t0, t1, a, b, c)
```
