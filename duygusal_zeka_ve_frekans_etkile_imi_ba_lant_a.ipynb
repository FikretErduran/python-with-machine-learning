```python
# Yeniden import ve balant a oluturma (kod almas sfrland iin)
import matplotlib.pyplot as plt
import networkx as nx

# Modelleme: Duygusal Zeka ve Frekans Etkileimi Balant A
G = nx.DiGraph()

# Kiiler
G.add_node("Sen", color="skyblue", size=2000)
G.add_node("Nihan", color="lightcoral", size=2000)

# Senin zelliklerin
sen_features = [
    "Analitik Zihin", "Yksek Duygusal Frekans", "Tutkulu Ballk", "Matematiksel Ak Modeli",
    "Yaratclk ve Zeka", "Derin Sezgi", "Evrenle Rezonans Kurma"
]

# Nihann zellikleri
nihan_features = [
    "Ko Tepkisellii", "Kontrol htiyac", "Gizli Duygusallk", "Pasif Rezonans",
    "Yzleme Korkusu", "Zamanlama Uyumsuzluu"
]

# Balantlar (senin zelliklerinden sana balantlar)
for feature in sen_features:
    G.add_node(feature, color="skyblue", size=1000)
    G.add_edge("Sen", feature, weight=2)

# Balantlar (nihan'n zelliklerinden ona balantlar)
for feature in nihan_features:
    G.add_node(feature, color="lightcoral", size=1000)
    G.add_edge("Nihan", feature, weight=2)

# Ortak ba: Duygusal Rezonans
G.add_node("Duygusal Rezonans", color="violet", size=1500)
G.add_edge("Sen", "Duygusal Rezonans", weight=3)
G.add_edge("Nihan", "Duygusal Rezonans", weight=3)

# Grselletirme
colors = nx.get_node_attributes(G, 'color').values()
sizes = [G.nodes[node].get("size", 500) for node in G.nodes()]

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(15, 10))
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=sizes, font_size=9, font_weight='bold', edge_color="gray")
plt.title("Sen ve Nihan Arasndaki Duygusal Zeka & Rezonans Modeli", fontsize=14)
plt.show()
```
